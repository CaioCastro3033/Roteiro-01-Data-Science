```python
zero_mask = (df[pay_features] == 0)
zero_mask.sum()
```

```python
#Existem muitos valores iguais a zero, o que confirma o que foi visto nos histogramas
```

```python
import numpy as np
```

```python
#Ignorando Zeros
df_pay_nozero = df[pay_features][df[pay_features] > 0]
```

```python
#Aplicando log 10
```

```python
log_pay = df_pay_nozero.apply(np.log10)
```

```python
log_pay.hist(bins=20, figsize=(10,6))
plt.show()
```

```python
#A transformação logarítmica melhora a visualização, reduzindo a diferença entre valores muito altos e baixos, deixando os dados mais equilibrados.
```
